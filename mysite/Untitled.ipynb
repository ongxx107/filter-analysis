```python
import matplotlib.pyplot as plt
import io
import urllib
import base64

plt.plot(range(10))
fig = plt.gcf()

buf = io.BytesIO()
#buf = io.StringIO()
fig.savefig(buf, format='png')
buf.seek(0)
string = base64.b64encode(buf.read())

uri = 'data:image/png;base64,' + urllib.parse.quote(string)
html = '<img src = "%s"/>' % uri
```
